## Model Selection

```python
# Assign the variables for traing the data and our target variable has price
X=df.drop(columns=['Price'])
y=df['Price']
```

```python
X=X.dropna()
y=y[X.index]
```

```python
X
```

```python
y
```

```python
# Train the model withe 80% 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
# import the necessary metrics regarding the regression models
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

## Linear Regression

```python
## import the onehotencoder change the categorical variable to numerical variable
# import the column transformer its change the columns wise categorical variable to numerical variable
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
categorical_columns=['Company','TypeName','Cpu brand','Gpu brand','OS_type']
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),categorical_columns)],remainder='passthrough')
```

```python
# import the pipline to establish the connection with model
# import the linear regression to train the model
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
lr=Pipeline(steps=[('preprocessor',ct),('model',LinearRegression())])
```

```python
# train the model with linear regression
lr.fit(X_train,y_train)
```

```python
# test the model
y_pred=lr.predict(X_test)
```

```python
# Evaluation the model
print('R2_Score:', round(r2_score(y_test,y_pred)*100,2))
print('MSE:', mean_squared_error(y_test,y_pred))
print('MAE:', mean_absolute_error(y_test,y_pred))
print('Train_Score:',round(lr.score(X_train,y_train)*100,2))
```

## Decision Tree Regressor

```python
# import the decision tree regressor and establish the connection withe dataset 
from sklearn.tree import DecisionTreeRegressor
dt=Pipeline(steps=[('preprocessor',ct),('model',DecisionTreeRegressor(max_depth=5))])
```

```python
# train the model with decisiontreeregressor
dt.fit(X_train,y_train)
```

```python
# Test the model
y_pred_dt=dt.predict(X_test)
```

```python
# evaluation the model 
print('R2_Score:',round(r2_score(y_test,y_pred_dt)*100,2))
print('MSE:', mean_squared_error(y_test,y_pred_dt))
print('MAE:', mean_absolute_error(y_test,y_pred_dt))
print('Train_Score:',round(dt.score(X_train,y_train)*100,2))
```

## Random Forest Regressor

```python
# import the randomforest regressor and establish the connection withe dataset 
from sklearn.ensemble import RandomForestRegressor
rf=Pipeline(steps=[('preprocessor',ct),('model',RandomForestRegressor(n_estimators=300,max_samples=0.8,max_features=0.75,max_depth=20))])
```

```python
# train the model
rf.fit(X_train,y_train)
```

```python
# test thee model
y_pred_rf=rf.predict(X_test)
```

```python
# Evaluation the model
print('R2_Score:', round(r2_score(y_test,y_pred_rf)*100,2))
print('MSE:', mean_squared_error(y_test,y_pred_rf))
print('MAE:', mean_absolute_error(y_test,y_pred_rf))
print('Train_Score:',round(rf.score(X_train,y_train)*100,2))
```

## KNN Regressor
