## Model Selection

```python
X=df.drop(columns=['Price'])
y=np.log(df['Price'])
```

```python
X
```

```python
y
```

```python
# Train the model 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
X_train = X_train.dropna()
y_train = y_train[X_train.index]  # Make sure y_train aligns with the new X_train
lr.fit(X_train, y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
print('R2_Score:', r2_score(y_test,y_pred))
print('MSE:', mean_squared_error(y_test,y_pred))
print('MAE:', mean_absolute_error(y_test,y_pred))
print('Train_Score:',lr.score(X_train,y_train))
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(random_state=44)
dt.fit(X_train,y_train)
```

```python
y_pred_dt=dt.predict(X_test)
```

```python
print('R2_Score:', r2_score(y_test,y_pred_dt))
print('MSE:', mean_squared_error(y_test,y_pred_dt))
print('MAE:', mean_absolute_error(y_test,y_pred_dt))
print('Train_Score:',dt.score(X_train,y_train))
```

## Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=300,random_state=0,max_samples=0.8,max_features=0.75,max_depth=20)
rf.fit(X_train,y_train)
```

```python
y_pred_rf=rf.predict(X_test)
```
